```python
#Imports
import pandas as pd
import numpy as np
#preprocessing tools
from sklearn.preprocessing import normalize
#KMeans clustering tools
from sklearn.metrics import silhouette_score,confusion_matrix
from sklearn.cluster import KMeans
#for visualization
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('data/wdms/WDMAll.csv',index_col=0)
wdm = df.iloc[:,1:]
wdm.head()
```

```python
km=KMeans(n_clusters=2,n_init='auto')
km.fit(wdm)
labels=(df['LABEL']=='Postive')
preds=km.predict(wdm)
score=silhouette_score(wdm,preds)
print('Sillhouette Score: '+str(score))
print('Confusion Matrix:')
print(confusion_matrix(labels,preds))
```
