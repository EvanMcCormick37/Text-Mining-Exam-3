```python
df = pd.read_csv('data/wdms/WDMAll.csv',index_col=0)
wdm=df.iloc[:,1:]
wdm.head()
```

```python
lda_model = LatentDirichletAllocation(n_components=2, 
                                     max_iter=100, learning_method='online')
lda_wdm=lda_model.fit_transform(wdm)
```

```python
vocab = list(wdm.columns)
vocab_array = np.asarray(vocab)

word_topic = np.array(lda_model.components_)
word_topic = word_topic.transpose()
```

```python
num_top_words = 20
fig,ax = plt.subplots(2,1,layout='constrained')

for i in range(2):
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title(label=('Topic #'+str(i+1)))
    top_words_idx = np.argsort(word_topic[:,i])[::-1]  # descending order
    top_words_idx = top_words_idx[:num_top_words]
    top_words = vocab_array[top_words_idx]
    top_words_shares = word_topic[top_words_idx, i]
    freq_df = pd.DataFrame(top_words_shares,index=top_words,columns=['freq'])
    ax[i].imshow(WordCloud().generate_from_frequencies(freq_df['freq']))
plt.show()
```
