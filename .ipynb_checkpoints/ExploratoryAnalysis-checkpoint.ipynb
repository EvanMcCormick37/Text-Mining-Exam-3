```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#read files in
df_all = pd.read_csv('data/wdms/WDMAll.csv',index_col=0)

wdm_pos = df_all[df_all['LABEL']=='Positive'].iloc[:,1:]
wdm_neg = df_all[df_all['LABEL']=='Negative'].iloc[:,1:]
wdm_neg.head()
```

```python
fig, ax = plt.subplots(2,1,constrained_layout=True)
titles=["Positive Reviews","Negative Reviews"]
for i, wdm in enumerate([wdm_pos,wdm_neg]):
    sums = wdm.sum()
    wordcloud = WordCloud().generate_from_frequencies(sums)
    ax[i].imshow(wordcloud)
    ax[i].set_title(titles[i])
```
